1Q) What is Plotly?
Ans : The plotly Python library is an interactive, open-source plotting library that supports over 40 unique chart types covering a wide range of statistical, financial, geographic, scientific, and 3-dimensional use-cases.

It is built on top of the Plotly JavaScript library (plotly.js), plotly enables Python users to create beautiful interactive web-based visualizations that can be displayed in Jupyter notebooks, saved to standalone HTML files, or served as part of pure Python-built web applications using Dash. The plotly Python library is sometimes referred to as "plotly.py" to differentiate it from the JavaScript library.

3Q) Plot a Pie Chart of Indian Car companies market share % values (max 6 Companies market share).

```python
Market_share = [24.5, 19, 16.5, 20.45, 7.6, 4.40]
companies = ["Maruti Suzuki", "Hyundai", "Kia", "Mahindra & Mahindra", "Tata Motors", "Toyota"]
plt.pie(Market_share, labels = companies)
plt.title("Market share of 4 wheeler Auto-companies")
```

4Q) Plot a Bar Chart with India's covid-19 Active cases ( Max 10 States )

```python
active_cases = [119867, 102523, 65335, 37526, 38178, 2687, 20170, 1357, 5787, 1402]
states = ["Maharashtra", "Kerala", "Karnataka", "Tamil Nadu", "Andhra Pradesh", "Uttar Pradesh", "West Bengal", "Delhi", "Chhattisgarh", "Rajasthan"]
plt.figure(figsize=(20, 10))
plt.title("Covid-19 Active cases in Top 10 states")
plt.bar(states, active_cases)
plt.ylabel("No. of active cases")
plt.xlabel(" Names of States")
```
