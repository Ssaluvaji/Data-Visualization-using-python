```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,10)
y = x*x
plt.plot(x,y,"g*--",linewidth =1,markersize= 10)
plt.title("Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
x = [1,2,3,4,5,6,7]
y = [160,140,140,145,175,165,180]
plt.plot(x,y,"gD--")
plt.title("Sales of the week",color="green")
plt.xlabel("Days of the week",color="green")
plt.ylabel("No. of Sales",color="green")
```
